```python
data="suurkorpus_2kki_nyyd_on_norm.txt"

with open(data, 'r') as file:
    file_content = file.read()

words=file_content.split()#
#words.remove("httpskrediidiskooreekrediidiskoorist","httpwwwrmkeeteemadloodusesliikujalekaitsealadkoikkaitsealad","")

print(len(words))

letters = {}

for word in words:
    for letter in word:
          if letter not in letters:
              letters[letter] = 1
          else:
              letters[letter] += 1

# To display the letter counts in appearing order
#for letter, count in letters.items():
#    print(f"{letter}: {count}")
longestword = 0
for word in words:
  word_length = len(word)
  if word_length > longestword:
    longestword = word_length
    print(word)
print(longestword)
```

```python
from collections import defaultdict, Counter
# Dictionary to hold letter counts per position
position_letter_counts = defaultdict(Counter)
position_totals = Counter()

for word in words:
    for pos, letter in enumerate(word):
        if letter.isalpha():  # Optional: skip punctuation/digits
            letter = letter.lower()  # Optional: make case-insensitive
            position_letter_counts[pos][letter] += 1
            position_totals[pos] += 1

# Convert to probabilities
position_letter_probs = {}

for pos, letter_counts in position_letter_counts.items():
    total = position_totals[pos]
    probs = {letter: count / total for letter, count in letter_counts.items()}
    position_letter_probs[pos] = probs

# Print results
for pos in sorted(position_letter_probs):
    print(f"Position {pos + 1}:")
    for letter, prob in sorted(position_letter_probs[pos].items()):
        print(f"  {letter}: {prob:.3f}")
```

```python
from collections import defaultdict

# Assume 'words' is a cleaned list of words (already split, lowercase, no punctuation)
letter_positions = defaultdict(list)

# Collect positions for each letter
for word in words:
    word = word.lower()
    for pos, letter in enumerate(word):
        if letter.isalpha():
            letter_positions[letter].append(pos)

# Compute mean position for each letter
mean_positions = {}

for letter, positions in letter_positions.items():
    mean_positions[letter] = sum(positions) / len(positions)

# Print mean positions
for letter in sorted(mean_positions):
    print(f"{letter}: {mean_positions[letter]:.2f}")
```

```python
from collections import defaultdict

# Clean the word list (no punctuation, lowercase)
words = [word.lower() for word in file_content.split() if word.isalpha()]

# Store all relative probabilities for each letter
letter_probabilities = defaultdict(list)

for word in words:
    L = len(word)
    if L < 2:
        continue  # skip 1-letter words to avoid division by zero

    for i, letter in enumerate(word):
        prob = i / (L - 1)  # normalized position: 0 to 1
        letter_probabilities[letter].append(prob)

# Compute mean probability for each letter
mean_position_probs = {
    letter: sum(probs) / len(probs)
    for letter, probs in letter_probabilities.items()
}

# Print the results
for letter in sorted(mean_position_probs):
    print(f"{letter}: {mean_position_probs[letter]:.3f}")
```
